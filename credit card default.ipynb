# overall outstanding amount trend

```python
# The outstanding amount is the current balance that indicates the total unpaid dues that you are liable to pay. It consists of various elements
# including balance transfers, online or offline purchases, cash advances, applicable fees, interest charges and any other charges.
# Outstanding = Bill Amount − amount paid

# Calculate outstanding amounts for each month
credit['outstanding_sep'] = credit['amt_bill_sep_2005'] - credit['amt_paid_sep_2005']
credit['outstanding_aug'] = credit['amt_bill_aug_2005'] - credit['amt_paid_aug_2005']
credit['outstanding_jul'] = credit['amt_bill_jul_2005'] - credit['amt_paid_jul_2005']
credit['outstanding_jun'] = credit['amt_bill_jun_2005'] - credit['amt_paid_jun_2005']
credit['outstanding_may'] = credit['amt_bill_may_2005'] - credit['amt_paid_may_2005']
credit['outstanding_apr'] = credit['amt_bill_apr_2005'] - credit['amt_paid_apr_2005']

# Sum outstanding amounts for the bank as a whole
overall_trend = {
    'April': credit['outstanding_apr'].sum(),
    'May': credit['outstanding_may'].sum(),
    'June': credit['outstanding_jun'].sum(),
    'July': credit['outstanding_jul'].sum(),
    'August': credit['outstanding_aug'].sum(),
    'September': credit['outstanding_sep'].sum()
}

# Convert to DataFrame for plotting
trend_credit = pd.DataFrame(list(overall_trend.items()), columns=['Month', 'Total Outstanding'])
trend_credit['Total Outstanding'] = trend_credit['Total Outstanding'] / 1e6  # in millions

# Plot the trend
plt.figure(figsize=(8,5))
plt.plot(trend_credit['Month'], trend_credit['Total Outstanding'], marker='o', color='blue')
plt.title('Overall Outstanding Amount Trend (Millions NT$)')
plt.xlabel('Month')
plt.ylabel('Outstanding Amount (Millions)')
plt.grid(True)
plt.show()
```

# Is there any relationship between in outstanding amount / trend with respect to age, education, marriage, credit limit

```python
#	Is there any relationship between in outstanding amount / trend with respect to age, education, marriage, credit limit
import seaborn as sns
from scipy.stats import f_oneway, pearsonr


# Calculate total outstanding per customer
credit['total_outstanding'] = (
     (credit['amt_bill_sep_2005'] - credit['amt_paid_sep_2005']) +
    (credit['amt_bill_aug_2005'] - credit['amt_paid_aug_2005']) +
    (credit['amt_bill_jul_2005'] - credit['amt_paid_jul_2005']) +
    (credit['amt_bill_jun_2005'] - credit['amt_paid_jun_2005']) +
    (credit['amt_bill_may_2005'] - credit['amt_paid_may_2005']) +
    (credit['amt_bill_apr_2005'] - credit['amt_paid_apr_2005'])
)

# 1 Correlation with Age & Credit Limit
corr_age, p_age = pearsonr(credit['AGE'], credit['total_outstanding'])
corr_credit, p_credit = pearsonr(credit['LIMIT_BAL'], credit['total_outstanding'])

print(f"Correlation (Age vs Outstanding): {corr_age:.4f}, p-value: {p_age:.4e}")
print(f"Correlation (Credit Limit vs Outstanding): {corr_credit:.4f}, p-value: {p_credit:.4e}")

# 2 ANOVA for Education
groups_edu = [group['total_outstanding'] for _, group in credit.groupby('EDUCATION')]
f_edu, p_edu = f_oneway(*groups_edu)
print(f"ANOVA (Education vs Outstanding): F={f_edu:.4f}, p-value={p_edu:.4e}")

# 3 ANOVA for Marriage
groups_marriage = [group['total_outstanding'] for _, group in credit.groupby('MARRIAGE')]
f_mar, p_mar = f_oneway(*groups_marriage)
print(f"ANOVA (Marriage vs Outstanding): F={f_mar:.4f}, p-value={p_mar:.4e}")

# 4 Visualization
plt.figure(figsize=(12, 8))

# Credit Limit vs Outstanding
sns.relplot(x='LIMIT_BAL', y='total_outstanding', data=credit, alpha=0.5)
plt.title('Credit Limit vs Outstanding Amount')
plt.show()
```

#  Does outstanding amount / trend affect the default behavior in next month.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind



credit['outstanding_sep'] = credit['amt_bill_sep_2005'] - credit['amt_paid_sep_2005']

# Separate into default groups
default_group = credit[credit['default payment next month'] == 1]['outstanding_sep']
non_default_group = credit[credit['default payment next month'] == 0]['outstanding_sep']

# T-test
t_stat, p_val = ttest_ind(default_group, non_default_group, equal_var=False)

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_val:.4e}")
if p_val < 0.05:
    print(" Outstanding amount is significantly different between defaulters and non-defaulters.")
else:
    print(" No significant difference in outstanding amount between defaulters and non-defaulters.")

# Visualization
plt.figure(figsize=(8,5))
sns.lineplot(x = 'default payment next month', y = 'outstanding_sep', data=credit, palette='coolwarm')
plt.title("Outstanding Amount (Sept) vs Default Status")
plt.xlabel("Default Next Month (0=No, 1=Yes)")
plt.ylabel("Outstanding Amount")
#plt.ylim(0, credit['outstanding_sep'].quantile(0.95))  # limit extreme outliers
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

credit['OUTSTANDING_BUCKET'] = pd.qcut(credit['amt_bill_sep_2005'] - credit['amt_paid_sep_2005'], q=4, labels=['Low','Med-Low','Med-High','High'])
ct = pd.crosstab(credit['OUTSTANDING_BUCKET'], credit['default payment next month'])
chi2, p, dof, exp = chi2_contingency(ct)
print("Chi-Square Test (Outstanding Bucket vs Default):")
print("Chi2:", chi2, "P-value:", p)

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
for i in range(1, 1):  # first 3 months as example
    sns.kdeplot(credit[credit['default payment next month']==0][f'amt_bill_sep_2005{i}'] - credit[credit['default payment next.month']==0][f'amt_paid_sep_2005{i}'],
                label=f'No Default - OUT{i}')
    sns.kdeplot(credit[credit['default payment next month']==1][f'amt_bill_sep_2005{i}'] - credit[credit['default payment next month']==1][f'amt_paid_sep_2005{i}'],
                label=f'Default - OUT{i}', linestyle="--")
plt.title("Outstanding Distributions (Default vs Non-Default)")
plt.xlabel("Outstanding Amount")
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

#  Flag overpayments for each month
bill_cols = ['amt_bill_sep_2005', 'amt_bill_aug_2005', 'amt_bill_jul_2005', 'amt_bill_jun_2005', 'amt_bill_may_2005',
             'amt_bill_apr_2005']
pay_cols  = ['amt_paid_sep_2005',  'amt_paid_aug_2005',  'amt_paid_jul_2005',  'amt_paid_jun_2005',  'amt_paid_may_2005',  'amt_paid_apr_2005']

for bill_col, pay_col in zip(bill_cols, pay_cols):
    credit[f'overpaid_{bill_col}'] = credit[pay_col] > credit[bill_col]

#  Flag if overpaid in any month
credit['overpaid_any_month'] = credit[[f'overpaid_{col}' for col in bill_cols]].any(axis=1)

#  Calculate default rate for overpayers vs non-overpayers
default_rates = credit.groupby('overpaid_any_month')['default payment next month'].mean() * 100
counts = credit.groupby('overpaid_any_month')['default payment next month'].count()

print("Default Rates (%):")
print(default_rates)
print("\nNumber of Customers in Each Group:")
print(counts)

#  Chi-square test (check if difference is significant)
contingency_table = pd.crosstab(credit['overpaid_any_month'], credit['default payment next month'])
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"\nChi-square Test: chi2={chi2:.4f}, p-value={p:.4e}")
if p < 0.05:
    print(" Significant difference in default rates between overpayers and non-overpayers.")
else:
    print(" No significant difference in default rates between overpayers and non-overpayers.")

# Visualization
default_rates.plot(kind='bar', color=['green', 'red'], edgecolor='black')
plt.title('Default Rate by Overpayment Status')
plt.ylabel('Default Rate (%)'),
plt.xticks([0, 1], ['No Overpayment', 'Overpaid'])
plt.grid(axis='y')
plt.show()
```
